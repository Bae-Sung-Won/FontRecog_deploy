```python
plt.rc('font', family='NanumGothic')
```

```python
classes = [i+1 for i in range(12)]
mse = [i for i in range(12)]
cos = [i for i in range(12)]
```

```python
df = pd.DataFrame({'클래스': classes, 'mse': mse, 'cos':cos}) 
```

```python
fig, ax = plt.subplots(1, 2, figsize=(18,8))
palette = sns.color_palette("bright", 12)

plt.subplot(1, 2, 1)
sns.barplot(data=df, x='클래스', y='mse',  palette=palette, dodge=False)
plt.title('MSE')

plt.subplot(1, 2, 2)
sns.barplot(data=df, x='클래스', y='cos', palette=palette, dodge=False)
plt.title('Cosine Similarity')
```
